```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,accuracy_score
from keras.layers import Input, Dense
from keras.models import Model
ratings_df = pd.read_csv('ratings.csv')

user_item_matrix = ratings_df.pivot(index='userId', columns='movieId', values='rating').fillna(0)

data = user_item_matrix.values

train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)

n_visible = train_data.shape[1]
n_hidden = 100

visible_input = Input(shape=(n_visible,), name='visible_input')
hidden_layer = Dense(n_hidden, activation='relu', name='hidden_layer')(visible_input)
visible_output = Dense(n_visible, activation='sigmoid', name='visible_output')(hidden_layer)

rbm = Model(visible_input, visible_output, name='rbm')

rbm.compile(optimizer='adam', loss='binary_crossentropy')

rbm.fit(train_data, train_data, epochs=10, batch_size=64, validation_data=(test_data, test_data))

predictions = rbm.predict(test_data)

mse = mean_squared_error(test_data, predictions)
print("Mean Squared Error:", mse)
```
